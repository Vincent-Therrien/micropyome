```python
import pandas as pd
from micropyome.taxa import regression
```

```python
DATASETS = [
    "../data/averill/bacteria",
    "../data/averill/fungi"
]

for dataset in DATASETS:
    print(dataset)
    for level in regression.TAXONOMIC_LEVELS:
        y = pd.read_csv(f"{dataset}/{level}/observed.csv")
        y = y.drop(y.columns[0], axis=1)  # Drop the ID of the sample.
        print(level)
        print(y.mean())
        print()
    print()
```

```python
dataset = "../data/neon/bacteria"
print(dataset)
for level in regression.TAXONOMIC_LEVELS:
    if level == "fg":
        continue
    y = pd.read_csv(f"{dataset}/Observed/{level}.csv", delimiter=";")
    y = y.drop(y.columns[0], axis=1)  # Drop the ID of the sample.
    print(level)
    print(y.mean())
    print()
print()
```

```python
dataset = "../data/neon/fungi"
print(dataset)
for level in regression.TAXONOMIC_LEVELS:
    y = pd.read_csv(f"{dataset}/Observed/{level}_core_observed.csv")
    y = y.drop(y.columns[0], axis=1)  # Drop the ID of the sample.
    print(level)
    print(y.mean())
    print()
print()
```

```python
fgs = [
    "assim_nitrate_reduction",
    "assim_nitrite_reduction",
    "cellulolytic",
    "chitinolytic",
    "copiotroph",
    "denitrification",
    "dissim_nitrate_reduction",
    "dissim_nitrite_reduction",
    "lignolytic",
    "methanotroph",
    "n_fixation",
    "nitrification",
    "oligotroph",
]

amounts = {}
for fg in fgs:
    y = pd.read_csv(f"../data/neon/bacteria/Observed/{fg}.csv", delimiter=";")
    y = y.drop(y.columns[0], axis=1)  # Drop the ID of the sample.
    print(fg)
    print(y.mean())
    amounts[fg] = y.mean()[fg]

for fg, a in amounts.items():
    print(f"{fg}: {a:.6f}")
```
